```python
import csv

nonCLR_folder = 'resnet18_MNIST_CEL_lr0.01_nonCLR_batch100_epochs100_xEpochs'
clr_folder = 'resnet18_MNIST_CEL_lr0.01-0.1_CLR_batch100_epochs100_xEpochs'
batch_size = 100
len_X_train = 60000
len_X_test = 10000

with open(nonCLR_folder + '/train_loss') as f:
    nonCLR_train_helper = []
    for row in csv.reader(f):
        nonCLR_train_helper.append(row)

with open(nonCLR_folder + '/val_loss') as f:
    nonCLR_val_helper = []
    for row in csv.reader(f):
        nonCLR_val_helper.append(row)

with open(clr_folder + '/train_loss') as f:
    clr_train_helper = []
    for row in csv.reader(f):
        clr_train_helper.append(row)

with open(clr_folder + '/val_loss') as f:
    clr_val_helper = []
    for row in csv.reader(f):
        clr_val_helper.append(row)
        
nonCLR_train = list(map(float, nonCLR_train_helper[0]))
nonCLR_val = list(map(float, nonCLR_val_helper[0]))
clr_train = list(map(float, clr_train_helper[0]))
clr_val = list(map(float, clr_val_helper[0]))
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot(np.array(clr_train) * batch_size / len_X_train, label='CLR training loss')
plt.plot(np.array(clr_val) * batch_size / len_X_test, label='CLR validation loss')
plt.plot(np.array(nonCLR_train) * batch_size / len_X_train, label='fixed rate training loss')
plt.plot(np.array(nonCLR_val) * batch_size / len_X_test, label='fixed rate validation loss')
plt.xlabel('Epochs')
plt.ylabel('Mean Cross Entropy Loss')
plt.legend()
```

```python
start = 10

plt.plot(np.array(clr_train) * batch_size / len_X_train, label='CLR training loss')
plt.plot(np.array(nonCLR_train) * batch_size / len_X_train, label='fixed rate training loss')
plt.ylim(0.0, 0.001)
plt.xlabel('Epochs')
plt.ylabel('Mean Cross Entropy Loss')
plt.legend()
```

## Plotting Accuracy after 3 Epochs

```python
with open('val_accuracy') as f:
    val_accuracy_helper = []
    for row in csv.reader(f):
        val_accuracy_helper.append(row)
        
val_accuracy = list(map(float, val_accuracy_helper[0]))
```

```python
plt.plot(val_accuracy)
print(np.argmax(np.array(val_accuracy)), max(val_accuracy))
```

## Checking if Batch-Size is important

```python
adam_folder = 'resnet18_MNIST_CEL_Adam_lr0.001_nonCLR_batch100_epochs100_xEpochs'

with open(adam_folder + '/train_loss') as f:
    adam_train_helper = []
    for row in csv.reader(f):
        adam_train_helper.append(row)

with open(adam_folder + '/val_loss') as f:
    adam_test_helper = []
    for row in csv.reader(f):
        adam_test_helper.append(row)
        
adam_train = list(map(float, adam_train_helper[0]))
adam_test = list(map(float, adam_test_helper[0]))
```

```python
plt.plot(np.array(clr_train) * batch_size / len_X_train, label='CLR training loss')
plt.plot(np.array(clr_val) * batch_size / len_X_test, label='CLR validation loss')
plt.plot(np.array(nonCLR_train) * batch_size / len_X_train, label='fixed rate training loss')
plt.plot(np.array(nonCLR_val) * batch_size / len_X_test, label='fixed rate validation loss')
plt.plot(np.array(adam_train) * batch_size / len_X_train, label='Adam training loss')
plt.plot(np.array(adam_test) * batch_size / len_X_test, label='Adam validation loss')
plt.xlabel('Epochs')
plt.ylabel('Mean Cross Entropy Loss')
plt.legend()
```
